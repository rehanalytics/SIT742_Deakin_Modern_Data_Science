#### 4.1 Defining and Calling Functions

A function definition starts with `def`, followed by the function name, parentheses, and a colon. The function body is indented.


```python
def say_hello():
    print('Hello world!')

say_hello()
say_hello()
```

A function can call another function. Here, `hello_threetimes()` reuses `say_hello()` instead of repeating the same print statement manually.


```python
def say_hello():
    print('Hello world!')

def hello_threetimes():
    say_hello()
    say_hello()
    say_hello()

hello_threetimes()
print('Done!')
```

Parameters let a function receive input values. The value supplied when the function is called is called an argument.


```python
abs(5)
```

```python
abs(-5)
```

```python
a = 30
b = 55
abs(a - b)
```

```python
pow(2, 3)
```

```python
pow(7, 4)
```

```python
def print_twice(param):
    print(param + ' ' + param)

print_twice('Hi')
print_twice('Hi' * 3)
```

Variables created inside a function are local to that function. The original example tried to print `cat` after the function finished; this is guarded below so you can see the expected error message without stopping the notebook.


```python
def cat_twice(part1, part2):
    cat = part1 + part2
    print(cat)
    print(cat)

str1 = 'Good'
str2 = 'Morning'
cat_twice(str1, str2)

try:
    print(cat)
except NameError as error:
    print('Expected scope error:', error)
```

Many functions return a value. Returning a value is different from printing it: a returned value can be assigned to a variable or passed into another expression.


```python
bigger = max(3, 5)
bigger
```

```python
x = abs(3 - 11) + 10
x
```

```python
def area(radius):
    temp = 3.14159 * radius ** 2
    return temp

print(area(3))
```

If a function reaches the end without a `return` statement, Python returns `None`. In this example, `absolute_value(0)` returns `None` because neither condition is true.


```python
def absolute_value(x):
    if x < 0:
        return -x
    elif x > 0:
        return x

print(absolute_value(0))
type(None)
```

Docstrings describe what a function does. They are available through `help()` and through the function's `__doc__` attribute.


```python
def cat_twice(part1, part2):
    '''Concatenates part1 and part2, and prints the result twice.'''
    cat = part1 + part2
    print(cat)
    print(cat)

help(cat_twice)
```

#### 4.2 Importing and Using Modules

A module contains related functions and values. The `math` module is part of the Python standard library.


```python
import math

math.pi
```

The original notebook showed these `math` examples with interactive prompt markers such as `>>>`. In a notebook, the expressions should be ordinary Python code cells.


```python
print(math.floor(3.4))   # Largest integer less than or equal to 3.4         #floor() means: Round a number down to the nearest whole integer.
print(math.floor(-3.4))
print(math.sqrt(5))
print(math.pow(2, 4))
```

If you import the module with `import math`, you access names with the module prefix, such as `math.pi`. Trying to use `pi` directly before importing it is an expected namespace error.


```python
import math
print(math.pi)

try:
    print(pi)
except NameError as error:
    print('Expected namespace error:', error)
```

You can import a specific object from a module when you want to use it directly.


```python
from math import pi

print(pi)
```

You may also import everything from a module. This can be concise, but it can make code harder to read because names appear without their module prefix.


```python
from math import *

cos(pi)
```

#### 4.3 Classes, Objects, and Methods

A class defines a new type. An object is an instance of that type. The simplest class body can use `pass` as a placeholder.


```python
class Person:
    pass    # An empty block

p = Person()
print(p)
```

Methods are functions defined inside a class. The first parameter is conventionally named `self`, which refers to the object receiving the method call.


```python
class Person:
    def say_hi(self):
        print('Hi, how are you doing?')

p = Person()
p.say_hi()
```

The `__init__` method runs when a new object is created. It is commonly used to initialise object attributes.


```python
class Person:
    def __init__(self, name):
        self.name = name

    def say_hi(self):
        print('Hi, ' + self.name + ', how are you doing?')

p = Person('James Bond')
p.say_hi()
```

Class variables are shared by all instances of a class. Object variables, such as `self.name`, belong to each individual object.


```python
class Robot:
    """Represents a robot, with a name."""
    population = 0

    def __init__(self, name):
        self.name = name
        print('(Initializing %s)' % (self.name))
        Robot.population += 1

    def die(self):
        print('%s is being destroyed!' % self.name)
        Robot.population -= 1
        if Robot.population == 0:
            print('%s was the last one.' % self.name)
        else:
            print('There are still %d robots working.' % Robot.population)

    def say_hi(self):
        print('Greetings, my masters call me %s' % self.name)

    @classmethod
    def how_many(cls):
        print('We have %d robots.' % cls.population)

droid1 = Robot('R2-D2')
droid1.say_hi()
Robot.how_many()
droid2 = Robot('C-3PO')
droid2.say_hi()
Robot.how_many()

print('\nRobots can do some work here.\n')
print("Robots have finished their work. So let's destroy them.")
droid1.die()
droid2.die()
Robot.how_many()
print(Robot.__doc__)
```

Inheritance lets a new class reuse and specialise behaviour from an existing class. The base class below stores the common attributes for school members, while the subclasses add teacher-specific and student-specific details.


```python
class SchoolMember:
    '''Represents any school member.'''
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print('(Initialized SchoolMember: %s)' % self.name)

    def tell(self):
        print('Name: %s Age: %d' % (self.name, self.age))

class Teacher(SchoolMember):
    '''Represents a teacher.'''
    def __init__(self, name, age, salary):
        SchoolMember.__init__(self, name, age)
        self.salary = salary
        print('(Initialized Teacher: %s)' % self.name)

    def tell(self):
        SchoolMember.tell(self)
        print('Salary:  %d ' % self.salary)

class Student(SchoolMember):
    '''Represents a student.'''
    def __init__(self, name, age, marks):
        SchoolMember.__init__(self, name, age)
        self.marks = marks
        print('(Initialized Student: %s)' % self.name)

    def tell(self):
        SchoolMember.tell(self)
        print('Marks: %d' % self.marks)

t = Teacher('Dr. Zhu', 35, 50000)
s = Student('John Smith', 22, 75)
print('\n')
members = [t, s]
for member in members:
    member.tell()
```
