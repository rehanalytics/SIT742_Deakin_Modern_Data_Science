#### 4.2 Importing a Module

The most explicit import style is `import module_name`. You then use the module prefix when calling its functions.


````python
import math

x = 2
y1 = math.exp(x)
y2 = math.log(x)

print('e^{} is {} and log({}) is {}'.format(x, y1, x, y2))
````

You can import one object from a module when you only need that object.


````python
from math import exp

x = 2
y = exp(x)  # No need to write math.exp() here.

print('e^{} is {}'.format(x, y))
````

Aliases are useful when a module or function has a long name, or when a community convention is widely understood.


````python
from math import exp as myfun

x = 2
y = myfun(x)

print('e^{} is {}'.format(x, y))
````

Wildcard imports bring all public names from a module into the current namespace. This can be convenient in a short teaching example, but it is usually clearer to use explicit imports in larger notebooks.


````python
from math import *

x = 2
y1 = exp(x)
y2 = log(x)

print('e^{} is {} and log({}) is {}'.format(x, y1, x, y2))
````

The domain of arc cosine is `[-1, 1]`. Guard the calculation before calling `math.acos()` when the value may be outside the valid range.


````python
# Interactive version:
# x = input('Enter a real number between -1 and 1: ')

x = '0.5'
x = float(x)
if -1 <= x <= 1:
    y = math.acos(x)
    print('acos({}) = {}'.format(x, y))
else:
    print('Out of range')
````

#### 4.4 The `datetime` Module

The `datetime` module includes classes for time, dates, time differences, formatting, and parsing.


````python
import datetime

t = datetime.time(11, 21, 33)
print(t)
print('hour  :', t.hour)
print('minute:', t.minute)
print('second:', t.second)
print('microsecond:', t.microsecond)
print('tzinfo:', t.tzinfo)
````

````python
today = datetime.date.today()
print(today)
print('ctime:', today.ctime())
print('tuple:', today.timetuple())
print('ordinal:', today.toordinal())
print('Year:', today.year)
print('Mon :', today.month)
print('Day :', today.day)
````

You can create a new date with `replace()`, keeping other parts of the date unchanged.


````python
d1 = datetime.date(2013, 3, 12)
print('d1:', d1)

d2 = d1.replace(year=2015)
print('d2:', d2)
````

Using `replace()` is not the only way to calculate future or past dates. `timedelta` represents a duration that can be added to or subtracted from a date or datetime.


````python
today_dt = datetime.datetime.today()
print(today_dt)

tomorrow = today_dt + datetime.timedelta(days=1)
print(tomorrow)
````

The default string representation of a datetime object uses an ISO-style format. Use `strftime()` to create a formatted string, and `strptime()` to parse a string back into a datetime object.


````python
today_dt = datetime.datetime.today()
print('ISO     :', today_dt)
````

````python
str_format = '%a %b %d %H:%M:%S %Y'
s = today_dt.strftime(str_format)
print('strftime:', s)
````

````python
print(s)

d = datetime.datetime.strptime(s, str_format)
print(d)
print('strptime:', d.strftime(str_format))
````

````python
# Define a string variable s with value 07/03/2017.
s = '07/03/2017'
str_format = '%m/%d/%Y'

d = datetime.datetime.strptime(s, str_format)
print(d)
````
